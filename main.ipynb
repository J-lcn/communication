```python

dim = 10
model = Model(dim = dim)
model.set_objective(name = 'quad_fun1')
model.diff()
model.diff_second()

n_iter, eps = 20000, 10**(-5)
optimizer_factory = OptimizerFactory(model = model, eps = eps)
optimizer = optimizer_factory.create_method(method = 'BFGS')
optimizer.set_init(lb = 9, ub = 100)
time_start = time.time()
for k in range(n_iter):
    optimizer.search_direction(iter_time=k)
    if optimizer.is_stop():
        break
    optimizer.get_step()
    optimizer.update()
    optimizer.moniter(iter_time = k)
time_end = time.time()
print('time cost', time_end - time_start, 's')
    
```
